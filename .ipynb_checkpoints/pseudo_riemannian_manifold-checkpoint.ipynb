## Geodesic Example

```python
def f_fun(x):
    
    return jnp.array([x[0]**2, x[1]])-x[2]**2
```

```python
M = PseudoRiemannianManifold(f=f_fun)
```

```python
M.dim = 3
M.emb_dim = 2
```

```python
Geodesic = JAXOptimization(M=M,
                           extrinsic_batch_size=2,
                           intrinsic_batch_size=None,
                           init_fun=None,
                           lr_rate=0.01,
                           optimizer=None,
                           T=100,
                           max_iter=1000,
                           tol=1e-4,
                           seed=2712,
                          )
```

```python
z0 = jnp.zeros(3, dtype=jnp.float32)
zT = jnp.ones(3, dtype=jnp.float32)
```

```python
test = Geodesic(z0,zT)
```

```python
test[-1]
```

```python
jnp.linalg.norm(test[-2].reshape(-1))
```

```python
M.length(test[0])
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')

ax.plot(test[0][:,0], test[0][:,1], test[0][:,2])
```

```python
plt.plot(test[0][:,0], test[0][:,1])
```
